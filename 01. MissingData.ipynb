# Manejando Datos Extraviados

En este JNB se revisar√° el tema de manejo de datos extraviados.

```python
# modules we'll use
import pandas as pd
import numpy as np

# read in all our data
nfl_data = pd.read_csv("https://media.githubusercontent.com/media/GTD2101-ACD-GA-AD22/pandas-clean-ricardo-qm/main/data/NFL%20Play%20by%20Play%202009-2017%20(v4).csv")

# set seed for reproducibility
np.random.seed(0) 
```

```python
# look at the first five rows of the nfl_data file. 
# I can see a handful of missing data already!
nfl_data.head()
```

# Cuantificando los datos extraviados

Lo primero que se puede hacer es cuantificar, es decir contar o sumarizar, los datos extraviados de nuestro DataFrame.

```python
missing_values_count = nfl_data.isnull().sum()
```

```python
missing_values_count.iloc[0:10]
```

```python
total_cells=np.product(nfl_data.shape)
```

```python
total_missing = missing_values_count.sum()
```

```python
# percent of data that is missing

percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

# Eliminando Datos Extraviados

La forma m√°s r√°pia y no siempre limpia de atender los datos extraviados es la que se presenta a continuaci√≥n:

```python
# remove all the rows that contain a missing value
nfl_data.dropna()
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % nfl_data.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```
